### LifeSquare

```python
nan_life_square_ids = data.loc[data["LifeSquare"].isnull(),'Id'].values
```

```python
# считаем средний процент жилой площади

percent_life_square = data.loc[
    data['LifeSquare'].notnull(),'LifeSquare'].sum() / data.loc[data['Square'] > 10,
    'Square'
].sum()

percent_life_square
```

```python
data.loc[
    data["LifeSquare"].isnull(),
    'LifeSquare'
] = data['Square'] * percent_life_square

data.loc[data['Id'].isin(nan_life_square_ids)].head(5)
```

```python
data.loc[data["LifeSquare"] > data["Square"], 'LifeSquare'] = data['Square'] * percent_life_square
```

### Floor

```python
# дома где квартира находиться выше чем этажей в доме
small_floor_house_ids = data.loc[data['Floor'] > data['HouseFloor'], 'Id']
data.loc[data['Floor'] > data['HouseFloor'], 'HouseFloor'] = data['Floor']
data.loc[data['Id'].isin(small_floor_house_ids)].head(5)
```

### DistrictId

```python
data_mean_district_price = data.groupby(['DistrictId', 'Rooms'], as_index=False)[['Price']].mean().rename(columns={'Price': 'MeanPrice'})
```

```python
data = pd.merge(data, data_mean_district_price, on=['DistrictId', 'Rooms'], how='left');
```

### Healthcare_1

```python
data.loc[data['Healthcare_1'].isnull(), 'Healthcare_1'] = 0
```

### Prepare data

```python
data = pd.get_dummies(data)
excluded_features = ['Id','DistrictId', 'Price', 'Ecology_2', 'Ecology_3', 'Shops_2', 'MeanPrice']
feature_names = []
for feature_name in data.columns:
    if feature_name not in excluded_features:
        feature_names.append(feature_name)
```

```python
X = pd.DataFrame(data, columns=feature_names)
y = pd.DataFrame(data['Price'].values, columns=['price'])
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=42)
```

### LinearRegression

```python
lr = LinearRegression()
lr.fit(X_train, y_train);
```

```python
y_lr_pred = lr.predict(X_test)
```

```python
r2 = r2_score(y_test, y_lr_pred)
r2
```
